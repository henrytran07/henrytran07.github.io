## 2.0 Multi-class Classification

Neural Networks are often used to classify data. Examples are neural networks: 
* take in photo and classfy subjects in the photos as {dog, cat, horse, other}
* take in a sentence and classify the 'parts of speech' of its elements: {noun, verb, adjective, etc}

A network of this type will have multiple units in its final layer. Each output is associated with a category. When an input example is applied to the network, the output with the highest value is the category predicted. If the output is applied to a softmax function, the output of the softmax will provide proabilities of the input being in each category. 

```python
# make 4-class dataset for classfication
classes = 4 
m = 100 
centers = ([-5, 2], [-2, -2], [1, 2], [5, -2])
std=1.0 
X_train, y_train = make_blobs(n_samples=m, centers= centers, cluster_std=std, random_state=30)
```

Each dot represents a training example. The axis (x0, x1) are the inputs and the color represents the class the example is associated with. Once trained, the model will be presented with a new example (x0, x1) and will predict the class. 

```python
# show classes in data set
print(f"unique classes {np.unique(y_train)}")
# show how classes are represented
print(f"class representation {y_train[:10]}")
# show shapes of our dataset
print(f"shape of X_train: {X_train.shape}, shape of y_train: {y_train.shape}")
```

## 2.2 Model

This lab will use 2 layer network. Unlike the binary classification networks, this network has four outputs. Given an input example, the output with the highest value is the predicted class of the predicted class of the input. 

Below is an example of how to contruct this network in Tensorflow. Notice the output uses a `linear` rather than a `softmax` activation. While it is possible to include the softmax in the output layer. It is more numerically stable if linear outputs are passed to the loss function during training. If the model is used to predict probabilities, the softmax can be applied at that point. 

```python
tf.random.set_seed(1234)
model = Sequential (
    [
        Dense(2, activation = 'relu', name = 'L1'),
        Dense(4, activation = 'linear', name = 'L2')
    ]
)
```

```python
model.compile (
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = tf.keras.optimizers.Adam(0.01),
)

model.fit (
    X_train, y_train, 
    epochs = 200
)
```

```python
l1 = model.get_layer("L1")
W1, b1 = l1.get_weights()
```
