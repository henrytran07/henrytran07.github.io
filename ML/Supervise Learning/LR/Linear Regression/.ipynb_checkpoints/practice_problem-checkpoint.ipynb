# Problem Statement 
Let's use the same two data point as before - a house with 1000 square feet sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000

|Size (10000 sqft) | Price (10000s of dollars) |
| -----------------|---------------------------|
|1                 | 300                       |
|2                 | 500                       |

```python
import numpy as np
import math, copy 
import matplotlib.pyplot as plt
```

```python
X_train = np.array([1.0, 2.0]) 
y_train = np.array([300, 500])
```

```python
print(f"X_train shape: {X_train.shape}, X_train value: {X_train}")
print(f"y_train shape: {y_train.shape}, y_train value: {y_train}") 
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0]
    
    cost = 0
    for i in range(m): 
        fwb_i = x[i] * w + b
        cost = cost + (fwb_i - y[i])**2
    total_cost = 1 / (2*m) * cost 
    return total_cost
```

```python
def compute_gradient(x, y, w, b): 
    dj_dw = 0 
    dj_db = 0 
    m = x.shape[0] 
    for i in range(m): 
        fwb_i = x[i] * w + b
        dj_dw_i = (fwb_i - y[i]) * x[i]
        dj_db_i = fwb_i - y[i]
        dj_dw = dj_dw + dj_dw_i
        dj_db = dj_db + dj_db_i
    dj_dw = (1/m) * dj_dw 
    dj_db = (1/m) * dj_db

    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, w_int, b_init, alpha, cost_function, gradient_function, num_iters): 
    w = copy.deepcopy(w_init) 
    b = copy.deepcopy(b_init) 

    J_history = []
    p_history = []
    for i in range(num_iters): 
        dj_dw, dj_db = gradient_function(X, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i < 100000: 
            J_history.append(cost_function(X, y, w, b)) 
            p_history.append([w, b])

        if i%math.ceil(num_iters / 10) == 0: 
            print(  f"Iteration:{i:4}, Cost: {J_history[-1]:0.2e} ", 
                    f"dj_dw: {dj_dw:0.3e}, dj_db: {dj_db:0.3e}", 
                    f"w: {w}, b: {b}") 
    return w, b, J_history, p_history                 
```

```python
#initialize variables 
w_init = 0 
b_init = 0

iterations = 10000 
tmp_alpha = 1.0e-2

w_final, b_final, J_hist, p_hist = gradient_descent(X_train, y_train, w_init, b_init, tmp_alpha, compute_cost, compute_gradient, iterations)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 6))
ax1.plot(J_hist[:100])
ax2.plot(1000 + np.arange(len(J_hist[1000:])), J_hist[1000:])
ax1.set_title("Cost function vs iterations (start)"); ax2.set_title("Cost function vs iterations (end)") 
ax1.set_ylabel("Cost"); ax2.set_ylabel("Cost") 
ax1.set_xlabel("iterations"); ax2.set_xlabel("iterations") 
```

```python
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0]
    cost = 0
    for i in range(m): 
        f_wb = np.dot(x[i], w) + b 
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2*m) * cost 
    
    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    m, n = x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0

    for i in range(m): 
        err = (np.dot(x[i], w) + b) - y[i]
        for j in range(n): 
            dj_dw [j]= dj_dw[j] +  err * x[i, j]
        dj_db = dj_db + err 
    dj_dw = (1 / m) * dj_dw 
    dj_db = (1 / m) * dj_db 

    return dj_dw, dj_db 
```

```python
tmp_dj_dw, tmp_dj_db = compute_gradient(X_train, y_train, w_init, b_init)
print(f'dj_dw at intial w, b: {tmp_dj_dw}') 
print(f'dj_db at intial w, b: {tmp_dj_db}') 
```

```python
def gradient_descent(x, y, w_int, b_int, alpha, gradient_function, cost_function, num_iterations):
    J_history = []
    w = copy.deepcopy(w_int) 
    b = copy.deepcopy(b_int)

    for i in range(num_iterations): 
        dj_dw, dj_db = gradient_function(x, y, w_int, b_int)

        w = w - alpha * dj_dw 
        b = b - alpha * dj_db

        if i < 100000: 
            J_history.append(cost_function(x, y, w, b)) 
        if i % math.ceil(num_iterations / 10) == 0: 
            print(f"Iteration: {i:4d}, Cost: {J_history[-1]:8.2f}   ") 
    return w, b, J_history
```

```python
initial_w = np.zeros_like(w_init)
initial_b = 0. 

iterations = 1000
tmp_alpha = 5.0e-7

w_final, b_final, J_history = gradient_descent(X_train, y_train, initial_w, initial_b, tmp_alpha,
                                                compute_gradient, compute_cost, iterations)
print(f"b, w found by gradient descent: {b_final:0.2f}, {w_final} ") 
m,_ = X_train.shape
for i in range(m): 
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target_value: {y_train[i]}")
```
