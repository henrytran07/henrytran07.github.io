## Cost function with regularization

Intuition: 
* If we have a good degree of polynomial function, it is a pretty good fit while if we use higher degree, it could result in overfitting scenario
e.g: $w_1x + w_2x^{2} + b$ and $w_1x + w_2x^{2} + w_3x^{3} + w_4x^{4} + b$ make $w_3, w_4$ really small ($\approx 0$)

$$min \frac{1}{2m}\sum\limits_{i = 1}^{m}(f_{w, b}{\mathbf{w}^{(i)}} - y^{(i)})^{2} + 1000w_3^2 + 1000w_4^2$$

- The only case that we make the function small: $w_3$ and $w_4$ are small or we can make $w_3$ and $w_4$ have tiny contributions
- We can have simpler function, which are less likely to overfit
- We can penalize the other features to make 0

- We have n features from $w_1, w_2 .... w_{100}$ as well as $b$ parameter -> n features

$$J(\mathbf{w}, b) = \frac{1}{2m}\sum\limits_{i = 1}^{m}(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)})^{2} + \frac{\lambda}{2m}\sum\limits_{j = 1}^{n}w_j^2$$
- regularization parameters $\lambda > 0$
- We don't need to penalize parameter b if it is too large. The relative $\lambda$ is very important to keep $w_j$ small and fit data
- If $\lambda = 0$, it will have complex functions
- If $\lambda$ is extremely large, it will chose $w_1, w_2, w_3, w_4$ = 0, hence, $f(x) = b$, resulting in underfitting function.
- Increasing the regularization parameter lambda reduces overfitting by reducing the size of the parameters. For some parameters that are near zero, this reduces the effect of the associated features.

## Regularized linear regression

$$J(\mathbf{w}, b) = \frac{1}{2m}\sum\limits_{i = 1}^{m}(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)})^{2} + \frac{\lambda}{2m}\sum\limits_{j = 1}^{n}w_j^2$$

            Gradient descent

$$\begin{align*}
\text{repeat }\lbrace\newline
\;\;\; w_j &= w_j - \alpha\frac{\partial J(\mathbf{w}, b)}{\partial(w_j)}\newline 
\;\;\; b &= b - \alpha\frac{\partial J(\mathbf{w}, b)}{\partial{b}}\newline 
\rbrace
\end{align*}$$ 


$$\frac{\partial J(\mathbf{w}, b)}{\partial w_j} = \frac{1}{m}\sum\limits_{i = 1}^{m} (f_{\mathbf{w}, b}(\mathbf{x^{(i)}}) - y^{(i)})x_j^{(i)} + \frac{\lambda}{m}w_j$$

$$ \frac{\partial J(\mathbf{w}, b)}{\partial b}= \frac{1}{m}\sum\limits_{i = 1}^{m}(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)})$$

$$w_j = w_j - \alpha[\frac{1}{m}\sum\limits_{i = 1}^{m}[f_{\mathbf{w}, b}(\mathbf{x^{(i)}}) - y^{(i)}] + \frac{\lambda}{m}w_j]$$

$$w_j = w_j - \alpha\frac{\lambda}{m}w_j - \alpha\frac{1}{m}\sum\limits_{i = 1}^{m}(f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) - y^{(i)})x_j^{i}$$
- The latter term is an usual update in gradient descent while $w_j(1 - \alpha\frac{\lambda}{m})$ is to shrink $w_j$
