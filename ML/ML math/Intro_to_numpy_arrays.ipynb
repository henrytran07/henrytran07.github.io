## 1. Basics of Numpy

```python
import numpy as np 
```

```python
one_dimensional_arr = np.array([10, 12])
```

```python
a = np.array([1, 2, 3])
print(a)
```

```python
b = np.arange(3)
print(b)
```

```python
c = np.arange(1, 20, 3)
print(c)
```

```python
lin_spaced_arr = np.linspace(0, 100, 5)
print(lin_spaced_arr)
```

```python
ones_arr = np.ones(3)
print(ones_arr)
```

```python
rand_arr = np.random.rand(3)
print(rand_arr)
```

```python
two_dim_arr = np.array([[1, 2, 3], [4, 5,6]])
print(two_dim_arr)
```

```python
a = [1, 2, 3, 4, 5]
```

```python
sliced_arr = a[1:4]
print(sliced_arr)
```

```python
sliced_arr = a[:3]
print(sliced_arr)
```

```python
sliced_arr = a[::2]
print(sliced_arr)
```

```python
a1 = np.array([[1, 1],
             [2, 2]])

a2 = np.array([[3, 3], 
             [4, 4]])

vert_stack = np.vstack((a1, a2))
print(vert_stack)
```

```python
hort_stack = np.hstack((a1, a2))
print(hort_stack)
```

```python
empty_arr = np.empty(3)
print(empty_arr)
```

```python
A = np.array([
        [-1, 3], 
        [3, 2]])
b = np.array([7, 1])
```

```python
A_system = np.hstack((A, b.reshape(2, 1)))
print(A_system)
```

```python
d2 = np.linalg.det(A)
print(d2)
```

```python
x = np.linalg.solve(A, b)
print(x)
```

```python
A = np.array([
    [4, -3, 1], 
    [2, 1, 3], 
    [-1, 2, -5], 
], dtype = np.dtype(float))

b = np.array([-10, 0, 17], dtype = np.dtype(float))
```

```python
x = np.linalg.solve(A, b)
print(x)
```

```python
A = np.array([
    [4, 9, 9], 
    [9, 1, 6], 
    [9, 2, 3], 
])

B = np.array([
    [2, 2], 
    [5, 7], 
    [4, 4]
])
```

```python
np.matmul(A, B)
```

```python
A@B
```

```python
A_eig = np.linalg.eig(A) 
print(A_eig)
```
